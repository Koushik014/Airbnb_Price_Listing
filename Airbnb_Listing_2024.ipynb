```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude','longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```

```python
# Compute the correlation matrix for the selected columns
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()

# Extract only the correlations involving 'price'
price_corr = corr[['price']].sort_values(by='price', ascending=False)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data=price_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation of Price with Other Variables')
plt.show()
```
